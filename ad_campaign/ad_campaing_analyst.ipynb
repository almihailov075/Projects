Для поиска аномалий найдем сначала скользящее среднее и построим график для наглядности

## _5. Cреднее количество дней от даты создания рекламного клиента и первым запуском рекламного объявления_

```python
# Сначала соединим две таблицы
ads_full_data = ads_clients_data.merge(ads_data, how='inner', on='client_union_id', suffixes=('_client_data', '_ad_data'))
ads_full_data.head()
```

```python
# Получим колонку между датой запуска рекламного объявления и датой создания рекламного клиента.
ads_full_data['delta_days'] = ads_full_data.date_ad_data - ads_full_data.create_date
```

```python
# Сгруппируем по id клиента и найдем минимальное значение
ads_min = ads_full_data.groupby('client_union_id').delta_days.min()
```

```python
ads_min.mean().days
```

Среднее количество дней от даты создания рекламного клиента и первым запуском рекламного объявления равно 124 дням

## _6. Вычислить конверсию от создания рекламного кабинета в запуск первой рекламы_

Конверсию нам надо вычислить в течении не более 365 дней

```python
n_days = pd.Timedelta(365, unit='day')
```

```python
lt365 = ads_min.loc[ads_min < n_days]
```

```python
np.round(len(lt365)/ads_clients_data.client_union_id.nunique() * 100, 2)
```

Меньше одного процента от всех созданных рекламных кабинетов запускают свою первую рекламу в течение первого года.

## _7. Надо разобить наших клиентов по промежуткам от создания до запуска рекламного объявления._

```python
cut_bins= [0, 30, 90, 180, 365]
```

```python
cut_bins = [pd.to_timedelta(x, 'd') for x in cut_bins]
```

```python
ads_min_cut = pd.cut(ads_min, cut_bins, labels=['less than 30 days', '30-90 days', '90-180 days', 'more than 180 days'])
```

```python
sd = ads_min_cut.value_counts()
sd
```

```python
round(ads_min_cut.value_counts(normalize=True) * 100, 1)
```

Чуть более одного процента запускают рекламное объявление менее чем за 30 дней, основная масса рекламных объявлений запускается в промежутке от 90 до 180 дней - 47 %.
